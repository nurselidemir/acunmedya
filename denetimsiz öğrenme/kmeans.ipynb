```python
X = df[ ["Annual Income (k$)", "Spending Score (1-100)"] ].values # df -> array e çevir
```

### K-Means

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters = 4, random_state = 42)
model.fit(X)
```

```python
model.cluster_centers_
# hangi x, y noktasına merkez olarak davrandığını gösterir.
```

```python
y = model.predict(X)
```

```python
y
```

```python
plt.figure(figsize = (8,6))
plt.scatter(X[:,0], X[:,1], c = y)
plt.scatter(model.cluster_centers_[:,0], model.cluster_centers_[:,1], c ="red", s = 500, marker= "X")
```

```python
# Outlier --> aykırı değerler
# Dirsek(Elbow Method)
```

```python
# inertia -> parametre - her noktanın kendi merkezine uzaklığının kareleri toplamı
model.inertia_
```

```python
def elbow_method():
    inertias = []  # Initialize the inertias list
    for k in range(2, 11):  # Range can be adjusted for different numbers of clusters
        model = KMeans(n_clusters=k, random_state=42)
        model.fit(X)  # Assuming X is your input data
        inertias.append(model.inertia_)  # Append the inertia value to the list
        
    plt.figure(figsize=(8,6))
    plt.plot(range(2, 11), inertias) 
```

```python
elbow_method()
```

```python
# Silhouette Method --> kendi merkezimi dikkate alırken diğer merkezleri de alıyoruz
# kendi merkezine yakınlık? diğer merkezlere uzaklık?
```

```python
from sklearn.metrics import silhouette_score
```

```python
def silh_method():
    scores = []
    for k in range(2,11):
        kmeans = KMeans(n_clusters = k, random_state = 42)
        y = kmeans.fit_predict(X)
        scores.append(silhouette_score(X,y))
    plt.figure(figsize=(8,6))
    plt.plot(range(2,11),scores)
    
```

```python
silh_method()
```

##### 1.ödev -> Z Davies Bouldin Method araştırıp uygula

Davies-Bouldin Metodu (DBI), bir kümeleme algoritmasının başarımını değerlendiren bir metriktir.
Davies-Bouldin İndeksi, kümelerin:

İç tutarlılığı (her kümedeki noktalar merkeze ne kadar yakın?)
Birbirinden ayrışması (kümeler ne kadar uzak?)
gibi iki temel özelliği değerlendirir.

Her küme, diğer kümelere göre kıyaslanır.
Kümeler arası benzerlik ne kadar fazlaysa (yani iç içe geçmiş kümeler varsa), skor yüksek çıkar.
Skor ne kadar düşük olursa, kümeler o kadar iyi ayrılmıştır.

Düşük Davies-Bouldin skoru = İyi kümeleme
Yüksek skor = Kümeler fazla karışmış

```python
from sklearn.metrics import davies_bouldin_score
```

```python
def db_method():
    scores = []
    for k in range(2, 11):
        kmeans = KMeans(n_clusters=k, random_state=42)
        labels = kmeans.fit_predict(X)
        db_score = davies_bouldin_score(X, labels)
        scores.append(db_score)
    
    plt.figure(figsize=(8, 6))
    plt.plot(range(2, 11), scores, marker='o')
    plt.title("Davies-Bouldin Index vs Cluster Count")
    plt.xlabel("Number of Clusters")
    plt.ylabel("Davies-Bouldin Index")
    plt.grid(True)
    plt.show()
```

```python
db_method()
```

##### 2.ödev -> Scaling(özellik ölçekleme) nedir? Nasıl yapılır?

Özellik ölçekleme (Feature Scaling), makine öğrenimi algoritmalarının daha etkili ve 
doğru sonuçlar üretmesi için farklı büyüklüklerdeki sayısal özelliklerin aynı ölçek aralığına getirilmesi işlemidir.

Neden Kullanılır?

- Algoritmanın her özelliğe eşit önem vermesini sağlar.  
- Modelin daha hızlı ve stabil öğrenmesini destekler.  
- Sayısal kararsızlıkları (numerical instability) azaltır.  
- Aşırı büyük değerlerin domine etmesini engeller.

Özellik Ölçekleme Nasıl Kullanılır?
Numerik veriyi seç: Sadece sayısal sütunlarda uygulanır.

Skaler seç: Aşağıdakilerden biri tercih edilir:

MinMaxScaler() → 0-1 arası normalizasyon

StandardScaler() → Ortalaması 0, std sapması 1

RobustScaler() → Aykırı değerlere karşı dayanıklı

Fit ve transform et: Skaler objeyi veriye uygula.
