```python
import pandas as pd
import numpy as np
```

## read_csv: CSV dosyasını okur ve DataFrame'e dönüştürür.

```python
dataFrame = pd.read_csv("netflix_users.csv")
```

```python
dataFrame
```

## describe: DataFrame'in temel istatistiksel bilgilerini gösterir.

```python
dataFrame.describe
```

## isnull() : Bir DataFrame veya Series içindeki eksik (NaN) değerleri kontrol etmek için kullanılır.

```python
dataFrame.isnull()
```

## sum() : Sütunların veya satırların toplamını hesaplar.

```python
dataFrame.sum()
```

## iloc : Satır ve sütunlara indeks numarası ile erişmek için kullanılır.

```python
dataFrame.iloc[5]
```

## shape : DataFrame'in boyutunu (satır ve sütun sayısı) döndürür.

```python
dataFrame.shape
```

## apply() : bir fonksiyonu DataFrame'in satırlarına veya sütunlarına uygulamak için kullanılır.

```python
dataFrame['Age'] = dataFrame['Age'].apply(lambda x: x ** 2)
dataFrame
```

## duplicated() : DataFrame'deki tekrar eden (duplicate) satırları tespit etmek için kullanılır.

```python
dataFrame.duplicated()
```

## isin() : Bu fonksiyon, belirli bir sütunda belirli bir değere sahip satırların seçilmesine yardımcı olur.

```python
dataFrame['Country'].isin(["USA"])
```

```python
dataFrame['Country'].isin(["USA"]).sum()
```
