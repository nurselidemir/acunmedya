```python
import pandas as pd
```

```python
df = pd.read_csv("spam_ham_dataset.csv")
df.head(2)
```

```python
df = df[ ['text','label_num'] ]
df.head(2)
```

```python
import nltk
nltk.download("stopwords")
```

```python
# hediye -> HEDƒ∞YE ->
# Hediye, hediye
import string

ps = nltk.stem.porter.PorterStemmer()

def clean_text(text):
    text = text.lower() # B√ºt√ºn harfleri k√º√ß√ºk yap.

    text = text.translate(str.maketrans('','', string.punctuation)) # B√ºt√ºn noktalama i≈üaretlerini kaldƒ±r.

    # "Merhaba,D√ºnya."
    # ["merhaba","d√ºnya"]
    words = text.split() # default => Bo≈üluk Merhaba D√ºnya => ["Merhaba","D√ºnya"]   #Merhaba,d√ºnya,abc -> ["Merhaba","d√ºnya","abc"]

    words = [ word for word in words if not word.isdigit() ] # "500 TL" "4.5G" "3D" -> Yalnƒ±zca numerik kelimeler alƒ±nmayacak.
    # Stem -> Kelimeleri k√∂k√ºne indirgemek -> Running,Run -> RUN
    # Remove Stopwords -> The, is, at, on
    # Superhero - The Superhero
    words = [ ps.stem(word) for word in words if word not in nltk.corpus.stopwords.words('english') ]

    return " ".join(words)
```

```python
df.head(3)
```

```python
df['clean_text'] = df['text'].apply(clean_text)
```

```python
df.head(3)
```

```python
from sklearn.model_selection import train_test_split

X = df['clean_text']
y = df['label_num']
```

```python
#Vectorize -> metni numerize etmek.
from sklearn.feature_extraction.text import TfidfVectorizer # TFIDF vectorizer nasƒ±l √ßalƒ±≈üƒ±r?
```

```python
tfidf = TfidfVectorizer()
X_tfidf = tfidf.fit_transform(df['clean_text'])
X_tfidf
```

```python
# [ 
#  [ 0.8, 0.5, 0.3, 0 ,0 ,0],
#  [ 0.4, 1, 0.2, 0 , 0]
# ]

X_train, X_test, y_train, y_test = train_test_split(X_tfidf, y, test_size=0.3, random_state=42)
```

```python
# Naive Bayes -> Bayes teoremi
from sklearn.naive_bayes import MultinomialNB

model = MultinomialNB()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
# precision -> kesinlik
# recall -> duyarlƒ±lƒ±k

# f1 score -> 

# support -> veri sayƒ±sƒ± 0->1121 1->431
```

TF-IDF Vectorizer

Bir kelimenin bir d√∂k√ºmandaki √∂nemini √∂l√ßen ve metin verilerini sayƒ±sal vekt√∂rlere
d√∂n√º≈üt√ºren istatistiksel bir tekniktir.

TF : Belirli bir kelimenin belgede ka√ß kez ge√ßtiƒüini √∂l√ßer.
IDF : Bu kelimenin t√ºm belgelerde ne kadar ‚Äúaz‚Äù ge√ßtiƒüini hesaplar.

Nadir ama anlamlƒ± kelimeler y√ºksek skora sahip olur. Bu da spam filtreleme gibi g√∂revlerde fayda saƒülar.

```python
tfidf = TfidfVectorizer(
    stop_words='english',
    max_df=0.9,
    min_df=3,
    ngram_range=(1, 2),   # unigram + bigram
    sublinear_tf=True     # log(tf) uygular
)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

```

```python
from sklearn.metrics import accuracy_score

# Naive Bayes Modeli
nb_model = MultinomialNB()
nb_model.fit(X_train, y_train)
nb_pred = nb_model.predict(X_test)
nb_accuracy = accuracy_score(y_test, nb_pred)

# Logistic Regression Modeli
from sklearn.linear_model import LogisticRegression
lr_model = LogisticRegression(max_iter=1000)
lr_model.fit(X_train, y_train)
lr_pred = lr_model.predict(X_test)
lr_accuracy = accuracy_score(y_test, lr_pred)

# Kar≈üƒ±la≈ütƒ±rma tablosu ve yorum
print("### üéØ Model Skor Kar≈üƒ±la≈ütƒ±rmasƒ±\n")
print("| Model                | Accuracy   |")
print("|----------------------|------------|")
print(f"| Naive Bayes          | %{nb_accuracy * 100:.2f}       |")
print(f"| Logistic Regression  | **%{lr_accuracy * 100:.2f}**   |\n")

# A√ßƒ±klayƒ±cƒ± yorum
print("Logistic Regression modeli, TF-IDF vekt√∂rleri √ºzerinde daha y√ºksek doƒüruluk saƒülamƒ±≈ütƒ±r. "
      "√ñzellikle `ngram_range=(1,2)` ve `sublinear_tf=True` gibi iyile≈ütirmeler sayesinde, "
      "model daha anlamlƒ± kelime gruplarƒ±nƒ± √∂ƒürenebilmi≈ü ve sonu√ß olarak daha isabetli sƒ±nƒ±flandƒ±rma yapabilmi≈ütir.")
```
