```python
import pandas as pd
```

```python
df = pd.read_csv("spam_ham_dataset.csv")
df.head(2)
```

```python
df = df[ ['text','label_num'] ]
df.head(2)
```

```python
import nltk
nltk.download("stopwords")
```

```python
# hediye -> HEDİYE ->
# Hediye, hediye
import string

ps = nltk.stem.porter.PorterStemmer()

def clean_text(text):
    text = text.lower() # Bütün harfleri küçük yap.

    text = text.translate(str.maketrans('','', string.punctuation)) # Bütün noktalama işaretlerini kaldır.

    # "Merhaba,Dünya."
    # ["merhaba","dünya"]
    words = text.split() # default => Boşluk Merhaba Dünya => ["Merhaba","Dünya"]   #Merhaba,dünya,abc -> ["Merhaba","dünya","abc"]

    words = [ word for word in words if not word.isdigit() ] # "500 TL" "4.5G" "3D" -> Yalnızca numerik kelimeler alınmayacak.
    # Stem -> Kelimeleri köküne indirgemek -> Running,Run -> RUN
    # Remove Stopwords -> The, is, at, on
    # Superhero - The Superhero
    words = [ ps.stem(word) for word in words if word not in nltk.corpus.stopwords.words('english') ]

    return " ".join(words)
```

```python
df.head(3)
```

```python
df['clean_text'] = df['text'].apply(clean_text)
```

```python
df.head(3)
```

```python
from sklearn.model_selection import train_test_split

X = df['clean_text']
y = df['label_num']
```

```python
#Vectorize -> metni numerize etmek.
from sklearn.feature_extraction.text import TfidfVectorizer # TFIDF vectorizer nasıl çalışır?
```

```python
tfidf = TfidfVectorizer()
X_tfidf = tfidf.fit_transform(df['clean_text'])
X_tfidf
```

```python
# [ 
#  [ 0.8, 0.5, 0.3, 0 ,0 ,0],
#  [ 0.4, 1, 0.2, 0 , 0]
# ]

X_train, X_test, y_train, y_test = train_test_split(X_tfidf, y, test_size=0.3, random_state=42)
```

```python
# Naive Bayes -> Bayes teoremi
from sklearn.naive_bayes import MultinomialNB

model = MultinomialNB()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
# precision -> kesinlik
# recall -> duyarlılık

# f1 score -> 

# support -> veri sayısı 0->1121 1->431
```

TF-IDF Vectorizer

Bir kelimenin bir dökümandaki önemini ölçen ve metin verilerini sayısal vektörlere
dönüştüren istatistiksel bir tekniktir.

TF : Belirli bir kelimenin belgede kaç kez geçtiğini ölçer.
IDF : Bu kelimenin tüm belgelerde ne kadar “az” geçtiğini hesaplar.

Nadir ama anlamlı kelimeler yüksek skora sahip olur. Bu da spam filtreleme gibi görevlerde fayda sağlar.

```python
tfidf = TfidfVectorizer(
    stop_words='english',
    max_df=0.9,
    min_df=3,
    ngram_range=(1, 2),   # unigram + bigram
    sublinear_tf=True     # log(tf) uygular
)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

```

```python
from sklearn.metrics import accuracy_score

# Naive Bayes Modeli
nb_model = MultinomialNB()
nb_model.fit(X_train, y_train)
nb_pred = nb_model.predict(X_test)
nb_accuracy = accuracy_score(y_test, nb_pred)

# Logistic Regression Modeli
from sklearn.linear_model import LogisticRegression
lr_model = LogisticRegression(max_iter=1000)
lr_model.fit(X_train, y_train)
lr_pred = lr_model.predict(X_test)
lr_accuracy = accuracy_score(y_test, lr_pred)

# Karşılaştırma tablosu ve yorum
print("### 🎯 Model Skor Karşılaştırması\n")
print("| Model                | Accuracy   |")
print("|----------------------|------------|")
print(f"| Naive Bayes          | %{nb_accuracy * 100:.2f}       |")
print(f"| Logistic Regression  | **%{lr_accuracy * 100:.2f}**   |\n")

# Açıklayıcı yorum
print("Logistic Regression modeli, TF-IDF vektörleri üzerinde daha yüksek doğruluk sağlamıştır. "
      "Özellikle `ngram_range=(1,2)` ve `sublinear_tf=True` gibi iyileştirmeler sayesinde, "
      "model daha anlamlı kelime gruplarını öğrenebilmiş ve sonuç olarak daha isabetli sınıflandırma yapabilmiştir.")
```
