## AGE ile MEDV (ev fiyatları) ilişkisi

```python
plt.scatter(dataFrame['AGE'], dataFrame['MEDV'])
plt.xlabel('Age')
plt.ylabel('Median Value of Homes (MEDV)')
plt.title('Age vs Median Home Price')
plt.show()
```

## CRIM (suç oranı) ile MEDV (ev fiyatları) ilişkisi

```python
plt.scatter(dataFrame['CRIM'], dataFrame['MEDV'])
plt.xlabel('Crime Rate')
plt.ylabel('Median Value of Homes (MEDV)')
plt.title('Crime Rate vs Median Home Price')
plt.show()
```
