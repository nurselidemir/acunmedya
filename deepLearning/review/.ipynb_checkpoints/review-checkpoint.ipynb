```python
query = "Select * from users"
```

```python
import pandas as pd
```

```python
df = pd.read_sql(query, connection)
df.head(1)
connection.close()
```

```python
df.head(1)
```

```python
texts = df['review'].values
ratings = df['rating'].values
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(texts, ratings, test_size = 0.3, random_state = 42)
```

```python
# tokenization, tokenizer 
# token -> metnin nümerik hale çevrilmiş hali
```

```python
# !pip install tensorflow
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer

tokenizer = Tokenizer(num_words = 10000, oov_token = '<OOV>')
# tokenizer bilmediği(öğrenmediği) bir kelime ile karşılaştığında o kelimeye <OOV> ve buna karşılık gelen değeri verir.
tokenizer.fit_on_texts(X_train)
```

```python
tokenizer.word_index
# bad music -> [149, 147]
```

```python
X_train_seq = tokenizer.texts_to_sequences(X_train)
X_test_seq = tokenizer.texts_to_sequences(X_test)
```

```python
# padding -> bütün cümleleri aynı uzunluğa getirmek.
from tensorflow.keras.preprocessing.sequence import pad_sequences

X_train_pad = pad_sequences(X_train_seq, maxlen = 50, padding = "post")
X_test_pad = pad_sequences(X_test_seq, maxlen = 50, padding = "post")
# post -> sonuna boşluk eklemek , pre -> başına boşluk eklemek
```

```python
X_train_pad[0]
```

```python
# modeli kurmak
```

```python
import tensorflow as tf
```

```python
model = tf.keras.Sequential([
        tf.keras.layers.Embedding(input_dim = 10000, output_dim = 64, input_length = 50),
# kelimeleri vektörlere çeviren derin öğrenme katmanı
# verdiğim 10000 kelimelik hazneyi 64 lük vektörlere çevir (kelimeleri daha anlamlı hale getirmek için)
        tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(64)),
# Bidirectional -> cümleleri düz ve ters oku LSTM kullanarak
# LSTM -> Long Short-Term Memory
        tf.keras.layers.Dense(32, activation = 'relu'),
        tf.keras.layers.Dropout(0.4), # % 40 oranında rastgele nöronları yoksayar. overfitting riski azalır.
        tf.keras.layers.Dense(1, activation = 'linear') # çıkış katmanı
# 1 -> çıktı adedi (yani puan) 
])

model.compile(loss = 'mean_squared_error', optimizer = 'adam', metrics = ['mae']) # eğitim öncesi parametreler
```

```python
# Ödev1 -> Aktivasyon fonksiyonu nedir araştıralım? Farklı türleri nelerdir?
```

```python
# Ödev2 -> Backpropogation(Geri yayılma) nedir? Optimizer türleri nelerdir?
```

```python
history = model.fit(X_train_pad, y_train, epochs = 3, validation_data = (X_test_pad, y_test), batch_size = 3)
```
